```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v0 = 50  # initial velocity (m/s)
theta = np.radians(70)  # launch angle (degrees)

drag = .9  # drag coefficient (not used in this simple model)

t = np.linspace(0, 10, 100)

g = 9.81  # acceleration due to gravity (m/s^2)

v0x = v0 * np.cos(theta)  # horizontal component of velocity
v0y = v0 * np.sin(theta)  # vertical component of velocity
x0 = 0  # initial x position (m)
y0 = 0  # initial y position (m)

x = x0 + v0x*t*drag
y = y0 + v0y*t*drag - 0.5 * g * t**2
plt.plot(x, y)
plt.xlabel('x (m)')
plt.ylabel('y (m)')



x1 = x0 + v0x*t
y1 = y0 + v0y*t - 0.5 * g * t**2
plt.plot(x1, y1, color='black', alpha=.6, linestyle='--')


plt.show()
```
